# 标准线性回归

```python
import numpy as np

x = 2 * np.random.rand(100,1)
y = 4 + 3 * x + np.random.randn(100,1) #添加噪音
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y,'b.')
plt.xlabel('$x_l$',fontsize=18)
plt.ylabel('$y$',fontsize=18,rotation=0)
plt.axis([0,2,0,15])
plt.show()
```

```python
x_b = np.c_[np.ones((100,1)),x]
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
```

```python
theta_best
```

```python
x_new = [[0],[2]]
x_new_b = np.c_[np.ones((2,1)),x_new]
y_pred = x_new_b.dot(theta_best)
```

```python
y_pred
```

```python
plt.plot(x_new,y_pred,'r-')
plt.plot(x,y,'b.')
plt.axis([0,2,0,15])
plt.show()
```

# 使用梯度下降

梯度下降包括批量梯度下降、随机梯度下降、小批量梯度下降，下面先使用原始python方法实现梯度下降

```python
alpha = 0.01 #学习率
epochs = 1000 #迭代次数
features = x_b.shape[1]
rows = y.shape[0]
theta = np.random.randn(features,1)    
cost_max = np.inf

gradient_descent_theta = []
```

```python
for epoch in range(epochs):
    cal_y = x_b.dot(theta)
    error = cal_y - y  
    gradients = 2/rows * x_b.T.dot(error)
    theta = theta - alpha * gradients
    gradient_descent_theta.append(theta)
    if (epoch % 50) == 0:
        cost = 2/rows * np.sum(np.square(x_b.dot(theta)-y),axis=0)
        print('epoch:{},cost:{}'.format(epoch,cost))
```

```python
theta
```

```python
x_new_b.dot(theta)
```

```python
def plot_gradient_descent(theta,alpha):
    m = len(x_b)
    plt.plot(x,y,'b.')
    epochs = 1000
    for epoch in range(epochs):
        if epoch < 10:
            y_predict = x_new_b.dot(theta)
            style = 'b-' if epoch >0 else 'r--'
            plt.plot(x_new,y_predict,style)
        gradient = 2/m * x_b.T.dot(x_b.dot(theta) - y)
        theta = theta - alpha * gradient
    plt.xlabel('$x$',fontsize=18)
    plt.axis([0,2,0,15])
    plt.title(r'$\alpha={}$'.format(alpha),fontsize=16)
```

使用不同的学习率反映梯度变化情况，第一条红色虚线表示起点

```python
np.random.seed(42)
theta = np.random.randn(features,1)
plt.figure(figsize=(10,4))
plt.subplot(131)
plot_gradient_descent(theta,alpha=0.02)
plt.ylabel('$y$',fontsize=18,rotation=0)
plt.subplot(132)
plot_gradient_descent(theta,alpha=0.1)
plt.subplot(133)
plot_gradient_descent(theta,alpha=0.5)
plt.show()
```

# 随机梯度下降

由于算法的随机性质，它比批量梯度下降要不规则的多，成本函数将不再是缓缓降低直到最小值，
而是不段上上下下，算法停下来的参数值肯定是足够好的，但不是最优的，为了解决这个困扰，
需要逐步降低学习率

```python
def learningrate_scheduce(t):
    t0,t1 = 5,50
    return t0 / (t + t1)
```

```python
alpha = 1 #学习率
epochs = 1000 #迭代次数
features = x_b.shape[1]
rows = y.shape[0]
theta = np.random.randn(features,1)    
cost_max = np.inf 

stochastic_gradient_descent_theta = []
for epoch in range(epochs):
    for i in range(rows):
        if epoch == 0 and i < 20:
            style = 'b-' if i > 0 else 'r--'
            y_pred = x_new_b.dot(theta)
            plt.plot(x_new,y_pred,style)
        index = np.random.randint(rows)
        xi = x_b[index:index+1]
        yi = y[index:index+1]
        cal_y = xi.dot(theta)
        error = cal_y - yi 
        gradients = 2 * xi.T.dot(error)
        alpha = learningrate_scheduce(epoch * rows + i)
        cost = 2 * np.sum(np.square(error),axis=0)
        theta = theta - alpha * gradients
        stochastic_gradient_descent_theta.append(theta)
    if (epoch % 50) == 0:
        print('epoch:{},cost:{}'.format(epoch,cost))
        
plt.plot(x,y,'b.')                              
plt.xlabel("$x_1$", fontsize=18)                    
plt.ylabel("$y$", rotation=0, fontsize=18)           
plt.axis([0, 2, 0, 15])                                                        
plt.show()    
```

# 小批量梯度下降

```python
n_iterations = 50
minibatch_size = 20

np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

t0, t1 = 200, 1000
def learning_schedule(t):
    return t0 / (t + t1)

t = 0

minibatch_gradient_descent_theta = []
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(rows)
    x_b_shuffled = x_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, rows, minibatch_size):
        t += 1
        xi = x_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        gradients = 2/minibatch_size * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(t)
        theta = theta - eta * gradients
        minibatch_gradient_descent_theta.append(theta)
```

```python
theta_path_bgd = np.array(gradient_descent_theta)
theta_path_sgd = np.array(stochastic_gradient_descent_theta)
theta_path_mgd = np.array(minibatch_gradient_descent_theta)
```

```python
plt.figure(figsize=(10,6))
plt.plot(theta_path_sgd[:, 0], theta_path_sgd[:, 1], "r-s", linewidth=1, label="Stochastic")
plt.plot(theta_path_mgd[:, 0], theta_path_mgd[:, 1], "g-+", linewidth=2, label="Mini-batch")
plt.plot(theta_path_bgd[:, 0], theta_path_bgd[:, 1], "b-o", linewidth=3, label="Batch")
plt.legend(loc="upper left", fontsize=16)
plt.xlabel(r"$\theta_0$", fontsize=20)
plt.ylabel(r"$\theta_1$   ", fontsize=20, rotation=0)
plt.axis([2.5, 4.5, 2.3, 3.9])
plt.show()
```
