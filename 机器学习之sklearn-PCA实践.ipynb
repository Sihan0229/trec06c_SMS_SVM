# PCA类说明

## PCA类参数

<ul>
    <li><b>n_components</b>: int,float,None or string，
        这个参数可以帮我们指定希望PCA降维后的特征维度数目。最常用的做法是直接指定降维到的维度数目，此时n_components是一个大于等于1的整数
        当然，我们也可以指定主成分的方差和所占的最小比例阈值，让PCA类自己去根据样本特征方差来决定降维到的维度数，此时n_components是一个（0，1]之间的数，
        也可以将参数设置为”mle”, 此时PCA类会用MLE算法根据特征的方差分布情况自己去选择一定数量的主成分特征来降维，
        也可以用默认值，即不输入n_components，此时n_components=min(样本数，特征数)，即会保留所有特征</li>
    <li><b>copy</b>: </li>
    <li><b>whiten</b>: 判断是否进行白化。所谓白化，就是对降维后的数据的每个特征进行归一化，让方差都为1.对于PCA降维本身来说，一般不需要白化。如果你PCA降维后有后续的数据处理动作，可以考虑白化。默认值是False，即不进行白化</li>
    <li><b>svd_solver</b>: 即指定奇异值分解SVD的方法，由于特征分解是奇异值分解SVD的一个特例，一般的PCA库都是基于SVD实现的。有4个可以选择的值：{‘auto’, ‘full’, ‘arpack’, ‘randomized’}。randomized一般适用于数据量大，数据维度多同时主成分数目比例又较低的PCA降维，它使用了一些加快SVD的随机算法。 full则是传统意义上的SVD，使用了scipy库对应的实现。arpack和randomized的适用场景类似，区别是randomized使用的是scikit-learn自己的SVD实现，而arpack直接使用了scipy库的sparse SVD实现。默认是auto，即PCA类会自己去在前面讲到的三种算法里面去权衡，选择一个合适的SVD算法来降维。一般来说，使用默认值就够了</li>
    <li><b>tol</b>: 计算奇异值的容忍度</li>

<ul>
    <li><b>components_</b>: 返回具有最大方差的成分(它包含的主成分是水平向量)，第一个主成分即等于pca.components_.T[:,0]</li>

## PCA压缩示例

<b>需求</b>:对mnist数据进行降维

```python
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_mldata

mnist = fetch_mldata('MNIST Original')
features,labels=mnist['data'],mnist['target']
print('原特征维数:',features.shape)
#对features进行降维处理，保留解释方差比为0.95
pca = PCA(n_components=0.95)
features_reduce = pca.fit_transform(features)
print('降维后的特征维数:',features_reduce.shape)
#将降维后的数据还原
features_recovered = pca.inverse_transform(features_reduce)
print('还原后的特征维数:',features_recovered.shape)
```

通过上面发现，对MNIST数据集应用主成分分析，然后保留其方差的95%，原来每个实例的784个特征变得只有154个特征，这保留了绝大多数差异性的同时，数据集大小变为不到原始的20%,这是一个合理的压缩比

在PCA投影上允许投影的逆转换，也可以将缩小的数据集解压缩回784维数据集，当然，得到的并非原始的数据，因为投影时损失了一部分信息(5%被丢弃的方差)，但是它很大可能非常接近于原始数据，原始数据和重建数据之间的均方距离被称为重建误差

## 显示原始和经过压缩和解压缩之后的数字图像

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = matplotlib.cm.binary, **options)
    plt.axis("off")
```

```python
randon_index = np.random.randint(0,70000,50)
plt.figure(figsize=(9,9))
plt.subplot(211)
plot_digits(features[randon_index,:])
plt.subplot(212)
plot_digits(features_recovered[randon_index,:])
```

<b>通过上下两幅图可以看出图像质量由轻微损伤，但是数字基本完好无损</b>

# 增量PCA

PCA实现有个问题:它需要整个训练集都先存入内存，才能运行SVD算法
    
为了解决这个问题，sklearn提供了IncrementalPCA类实现增量PCA,可以将训练集分成一个个小批量，一次给IPCA算法喂一个。对于大型训练集来说，这个方法很有用，并且还可以在线应用PCA

## 示例

```python
from sklearn.decomposition import IncrementalPCA

n_batches = 100
inc_pca = IncrementalPCA(n_components = 154)
for x_batch in np.array_split(features,n_batches):
    inc_pca.partial_fit(x_batch)
    
features_incremental_reduce = inc_pca.transform(features)
print(features_incremental_reduce.shape)
```

# 练习

<b>需求:</b>
加载MNIST数据集，将其分为一个训练集和一个测试集(将前60000个实例用于训练，其余10000个用来测试)，在训练集上训练一个随机森林分类器，并记录训练时长，然后在测试集上对结果模型进行评估。接下来用PCA给数据降维，方差解释比为95%。在降维后的新数据集上训练一个新的随机森林分类器，看看需要多长时间，是不是快得多？最后，在测试集上评估分类器，跟前一个分类器比起来如何？

```python
from sklearn.datasets import fetch_mldata
from sklearn.model_selection import train_test_split

mnist = fetch_mldata('MNIST Original')
features,labels=mnist['data'],mnist['target']
print(features.shape)

# 拆分数据集
x_train,x_test,y_train,y_test = train_test_split(features,labels,test_size=10000)
print(x_train.shape,y_train.shape)
```

```python
#在训练集上训练一个随机森林分类器
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import time

start =time.time()
rfc_clf = RandomForestClassifier(n_estimators = 100,criterion='gini',bootstrap=True,verbose=1)
rfc_clf.fit(x_train,y_train)
y_pred = rfc_clf.predict(x_test)
end = time.time()
print('elapsed time:',end-start)
print('accuracy_score:',accuracy_score(y_test,y_pred))
```

```python
# 用PCA给数据降维，方差解释比为95%
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
X_reduce = pca.fit_transform(features)
print(X_reduce.shape)
```

```python
#重新划分数据集
x_reduce_train,x_reduce_test,y_reduce_train,y_reduce_test=train_test_split(X_reduce,labels,test_size=10000)
```

```python
rfc_clf2 = RandomForestClassifier(n_estimators = 100,criterion='gini',bootstrap=True,verbose=1)
start =time.time()
rfc_clf2.fit(x_reduce_train,y_reduce_train)
y_pred1 = rfc_clf2.predict(x_reduce_test)
end = time.time()
print('elapsed time:',end-start)
print('accuracy_score:',accuracy_score(y_reduce_test,y_pred1))
```
